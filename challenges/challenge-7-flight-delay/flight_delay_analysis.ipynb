# Flight Delay Prediction — Data Exploration & Model Building

## Objective

Build a machine learning model that predicts the probability that a flight will be delayed by more than 15 minutes, given a day of the week and a destination airport.

## Dataset

The dataset (`data/flights.csv`) contains information about US flights from 2013 sourced from the FAA. It includes carrier information, origin/destination airports, scheduled and actual times, and delay flags.

### Columns

| Column | Description |
|--------|-------------|
| Year | Year of the flight |
| Month | Month of the flight |
| DayofMonth | Day of the month |
| DayOfWeek | Day of the week (1=Monday, 7=Sunday) |
| Carrier | Airline carrier code |
| OriginAirportID | Numeric ID of the origin airport |
| OriginAirportName | Name of the origin airport |
| OriginCity | Origin city |
| OriginState | Origin state |
| DestAirportID | Numeric ID of the destination airport |
| DestAirportName | Name of the destination airport |
| DestCity | Destination city |
| DestState | Destination state |
| CRSDepTime | Scheduled departure time |
| DepDelay | Departure delay in minutes |
| DepDel15 | 1 if departure delay > 15 min, 0 otherwise |
| CRSArrTime | Scheduled arrival time |
| ArrDelay | Arrival delay in minutes |
| ArrDel15 | 1 if arrival delay > 15 min, 0 otherwise |
| Cancelled | 1 if flight was cancelled, 0 otherwise |
